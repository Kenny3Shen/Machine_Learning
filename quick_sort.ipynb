```python
import numpy as np
```

```python
arr = [np.random.randint(0,1000000) for i in range(1000000)]
arr
```

```python
def divide(left, right) -> int:
    pivot = arr[left]
    while left < right:
        while left < right and arr[right] >= pivot:
            right -= 1
        arr[left] = arr[right]
        
        while left < right and arr[left] <= pivot:
            left += 1
        arr[right] = arr[left]
    arr[left] = pivot
    return left
def conquer(left, right):
    if left < right:
        pivot = divide(left, right)
        conquer(left, pivot - 1)
        conquer(pivot + 1, right)
```

```python
conquer(0, len(arr) - 1)
```
